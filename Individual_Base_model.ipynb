```python
import numpy as np
import matplotlib.pyplot
```

```python
import random # Added import for the random module

#creating trait based on genotype for an individual

class Individual:
  def __init__(self, gene_loci): # Corrected _init_ to __init__
    self.gene_loci = gene_loci
    self.genotype = self.generate_genotypes()
    self.trait_z = self.calculate_trait_z() # Corrected typo in method call

  def generate_genotypes(self):
    """Generates random alleles (0, 1, or 2) for each locus"""
    #0: homozygote recessive, 1: heterozygotes, 2: homozygote dominate
    return [random.choice([0, 1, 2]) for _ in range(self.gene_loci)]

  def calculate_trait_z(self): # Corrected typo in method definition
    """Calculates Trait Z based on additive genetic effects."""
    #each allele adds 0.5 to the trait, 1 allele adds 0.5, 2 allele adds 1
    trait_value = sum(self.genotype) * 0.5
    return trait_value # Added return statement

population = [Individual(gene_loci=5) for _ in range(100)] # Corrected syntax here

for i, ind in enumerate(population):
  print(f"Individual {i+1}: Genotype {ind.genotype}, Trait z: {ind.trait_z}")
```

```python
import numpy as np # Added import for numpy

#assinging fitness based on Trait z

def calculate_fitness(trait_z, z_opt=5, selection_strength=1.0):
  """Calculates fitness based on a trait z using a Gaussian selection gradiant. Fitness is 1.0 at z_opt and decreases as z moves away."""
  # Ensure trait_z is a numpy array for vectorized operations
  trait_z = np.asarray(trait_z)
  fitness = np.exp(-(trait_z - z_opt)**2) / (2 * selection_strength**2)
  return fitness

# Extract trait_z values from the population of Individual objects
# and assign to initial_population_fitness
initial_population_fitness = calculate_fitness([ind.trait_z for ind in population])

print(initial_population_fitness)
```

```python
#next generation with asexual mating using poisson distrbution for number of offsring

base_rate = 2.0 #expected number of offspring
lambdas = base_rate * initial_population_fitness # Use the renamed variable

offspring_counts = np.random.poisson(lambdas)

print(offspring_counts)
```

```python
# Applying a fitness for offsrping same as parent (for asexual mating assuming no mutations ect)

parental_fitness = np.array(initial_population_fitness) # Use the renamed variable
offspring_counts = np.array(offspring_counts)

offspring_fitness = np.repeat(parental_fitness, offspring_counts)

print(offspring_fitness)
```

```python
num_generations = 10 # Example value, you might want to adjust this
current_fitness = offspring_fitness # Initialize current_fitness with the result from the previous step

#loop for multiple generations

for gen in range(num_generations):
  offspring_counts = np.random.poisson(lam=current_fitness)
  print(f"Gen {gen}: Offspring Counts = {offspring_counts}") # Print offspring counts

  #check for extinction
  if offspring_counts.sum() == 0:
    print(f"Population went extinct at generation {gen}")
    break # Exit the loop if population goes extinct

  current_fitness = np.repeat(current_fitness, offspring_counts)
  print(f"Gen {gen}: Current Fitness (new generation) = {current_fitness}") # Print fitness of the new generation
  print(f"Gen {gen}: Population size = {len(current_fitness)}")
```
